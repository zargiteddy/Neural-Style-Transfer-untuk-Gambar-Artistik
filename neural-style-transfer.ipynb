### 1. Import _libraries_ yang dibutuhkan

```python
import tensorflow_hub as hub
import tensorflow as tf
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model = hub.load('https://kaggle.com/models/google/arbitrary-image-stylization-v1/frameworks/TensorFlow1/variations/256/versions/1')
```

### 2. Lakukan praproses pada gambar

```python
def load_gambar(img_path):
    gambar = tf.io.read_file(img_path)
    gambar = tf.image.decode_image(gambar, channels = 3)
    gambar = tf.image.convert_image_dtype(gambar, tf.float32)
    gambar = gambar[tf.newaxis, :]
    return gambar
```

```python
gambar_sample = load_gambar('myphoto.png')
gambar_style = load_gambar('sketch.jpg')
```

### 3. Visualisasi

```python
gambar_sample.shape
```

```python
plt.imshow(np.squeeze(gambar_style))
plt.show()
```

### 4. Terapkan _style_ pada gambar sample

```python
gambar_final = model(tf.constant(gambar_sample), tf.constant(gambar_style))[0]
```

```python
plt.imshow(np.squeeze(gambar_final))
plt.show()
```
